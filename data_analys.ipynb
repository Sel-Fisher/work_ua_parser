# Saves parsed data

```python
import pandas as pd
from datetime import datetime
jobs = pd.read_csv("vacancies.csv")
```

# Graph statistic for Distribution of Experiences

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
sns.countplot(x="experience", data=jobs)
plt.title("Distribution of Experiences")
plt.xlabel("Experience Level")
plt.ylabel("Count")
timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
plt.savefig(f"data_analys_history/experience_distribution_{timestamp}.png")
plt.show()
```

# Graph statistic for Distribution of Technologies

```python
technologies = jobs["technologies"].str.split(",", expand=True).stack()
plt.figure(figsize=(12, 8))
technologies.value_counts().sort_values(ascending=False).plot(kind="bar")
plt.title("Distribution of Technologies")
plt.xlabel("Technology")
plt.ylabel("Count")
plt.savefig(f"data_analys_history/technologies_distribution_{timestamp}.png")
plt.show()
```

# Graph statistic for Experience vs. Number of Technologies

```python
import plotly.express as px
import plotly.io as pio

fig = px.scatter(jobs, x="experience", y=jobs["technologies"].str.count(",") + 1,
                 title="Experience vs. Number of Technologies",
                 labels={"experience": "Experience Level", "y": "Number of Technologies"})
pio.write_image(fig, f"data_analys_history/experience_vs_technologies_scatter_{timestamp}.png")
fig.show()
```
